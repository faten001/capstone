### Connect to the database

Let us first load the SQL extension and establish a connection with the database


```python
import csv, sqlite3

con = sqlite3.connect("my_data1.db")
cur = con.cursor()
```

```python
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.Date = pd.to_datetime(df.Date,format="%d/%m/%Y",errors='ignore')
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
```

```python
df.Mission_Outcome.isna().sum()
```

```python
result = cur.execute('select * from SPACEXTBL;')
columns = [i[0] for i in cur.description]
columns
```

## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes
   For Example "Landing_Outcome"**

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
cur.execute('select distinct(Launch_Site) from SPACEXTBL;')
cur.fetchall()
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
data = pd.DataFrame()
cur.execute("select * from SPACEXTBL where Launch_Site like 'CCA%' limit 5;")
for i in cur:
    data = data.append(pd.Series(i),ignore_index=True)
data.columns = columns
data
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
cur.execute("select SUM(PAYLOAD_MASS__KG_) from SPACEXTBL where Customer like 'NASA%';")
cur.fetchall()
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
cur.execute("select AVG(PAYLOAD_MASS__KG_) from SPACEXTBL where Booster_Version='F9 v1.1';")
cur.fetchall()
```

```python
cur.execute("select * from SPACEXTBL;")
cur.fetchall()
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
cur.execute("select min(Date) from SPACEXTBL where Landing_Outcome='Success (ground pad)';")
cur.fetchall()
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
# Success (drone ship)
cur.execute("select Booster_Version from SPACEXTBL where Landing_Outcome='Success (drone ship)' and (PAYLOAD_MASS__KG_>4000 and PAYLOAD_MASS__KG_<6000);")
cur.fetchall()
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
cur.execute("select Mission_Outcome,count(Mission_Outcome) from SPACEXTBL group by Mission_Outcome;")
cur.fetchall()
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
cur.execute("select Booster_Version from SPACEXTBL where PAYLOAD_MASS__KG_ = (select max(PAYLOAD_MASS__KG_) from SPACEXTBL);")
cur.fetchall()
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.**


```python
cur.execute("select substr(Date,6,2),Booster_Version,Launch_Site from SPACEXTBL where substr(Date,0,5) = '2015' and Landing_Outcome='Failure (drone ship)';")
cur.fetchall()
```

### Task 10




##### Rank the  count of  successful landing_outcomes between the date 04-06-2010 and 20-03-2017 in descending order.


```python
cur.execute("select Landing_Outcome,count(Landing_Outcome) as count from \
            (select * from SPACEXTBL where (Landing_Outcome like 'Success%' OR Landing_Outcome like 'Failure%') AND (Date Between '2010-06-04' AND '2017-03-20'))\
            group by Landing_Outcome order by count DESC;")
cur.fetchall()
```
